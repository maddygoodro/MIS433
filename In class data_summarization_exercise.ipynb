## DataFrame with columns

- year,month,day
        Date of departure    
- dep_time,arr_time
        Actual departure and arrival times (format HHMM or HMM), local tz.
- sched_dep_time,sched_arr_time
        Scheduled departure and arrival times (format HHMM or HMM), local tz.    
- dep_delay,arr_delay
        Departure and arrival delays, in minutes. Negative times represent early departures/arrivals.
- hour,minute
        Time of scheduled departure broken into hour and minutes.
- carrier
        Two letter carrier abbreviation. See airlines() to get name
- tailnum
        Plane tail number
- flight
        Flight number
- origin,dest
        Origin and destination. See airports() for additional metadata.
- air_time
        Amount of time spent in the air, in minutes
- distance
        Distance between airports, in miles
- time_hour
        Scheduled date and hour of the flight as a date. Along with origin, can be used to join flights data to weather data.

## Exercises

Write Python scripts to answer the following questions.

```python
# How many distinct airplanes are there in this dataset?
flights.tailnum.nunique()
```

```python
# Which hours of the day are the busiest?
# List the top 5 hours with the most flights.
flights.groupby('hour').size().sort_values(ascending=False).head(5)

#also works:
flights['hour'].value_counts().head(5)
```

```python
# Find all flights (identified by 'carrier' and 'flight') that always depart at least 60 minutes late.
#once we use group by we use min so later we can show anything greater than 60
df1 = flights.groupby(['carrier','flight']).agg({'dep_delay':'min'})
#save in an intermediary data frame so we can then filter by greater than 60
df1[df1.dep_delay>=60]
#flights.query('dep_delay >= 60').groupby(['carrier','flight']).size()
```

```python
flights.groupby(['carrier','flight']).agg({'dep_delay':'min'}).query('dep_delay>=60')
```

```python
# How many flights always depart at least 60 minutes late?
#the bracket with a 0 makes it return only the first element which is the only value we want as the result
flights.groupby(['carrier','flight']).agg({'dep_delay':'min'}).query('dep_delay>=60').shape[0]
```

```python
# Find all United Airlines (UA) flights (identified by 'carrier' and 'flight')
# that depart (at least) 60 minutes late.
flights[flights.carrier=='UA'].groupby(['carrier','flight']).agg({'dep_delay':'min'}).query('dep_delay>=60')
```

```python
# How many United Airlines (UA) flights depart (at least) 60 minutes late?
flights[flights.carrier=='UA'].groupby(['carrier','flight']).agg({'dep_delay':'min'}).query('dep_delay>=60').shape[0]
```

```python
# Which flights that always arrived on time in December?
#group by carrier and flight number to show specific flights
#show max arrival delay cuz if the max delay is =>0 then they're all equal to 0
flights[flights.month==12].groupby(['carrier','flight']).agg({'arr_delay':'max'}).query('arr_delay<=0')
#flights[flights.arr_delay<=0].groupby(['month','day']).size()
```

```python
# How many flights always arrived on time in December?
flights[flights.month==12].groupby(['carrier','flight']).agg({'arr_delay':'max'}).query('arr_delay<=0').shape[0]
```

```python
# Create a dataframe 'cancelled' for all cancelled flights.
# Assumption: flights that are not cancelled should have values for dep_time.
cancelled = flights[flights.dep_time.isna()]
cancelled.head()
```

```python
# Look at the number of cancelled flights per day.
# Find the top 10 days with the most cancelled flights.
cancelled[['year','month','day']].value_counts().head(10)

#alt way
cancelled.groupby(['year','month','day']).agg({'flight':'size'}).sort_values('flight',ascending=False).head(10)
```

```python
# Which carrier has the worst average arrival delay?
flights.groupby('carrier').agg({'arr_delay':'mean'}).sort_values('arr_delay')
```

```python
# Which plane (tailnum) has the worst on-time record?
df1 = flights.groupby('tailnum').agg({'arr_delay':['mean','min','max']})
#assign new names to the columns
df1.columns = ['ad_mean','ad_min','ad_max']
df1[df1.ad_mean.notna()].sort_values('ad_mean').tail(1)
```

```python
# What time (hour) of day should you fly if you want to avoid delays as much as possible?
df1 = flights.groupby('hour').agg({'dep_delay':['size','min','max']})
df1
```

```python
# For each destination, compute the total minutes of arrival delay.
flights.groupby('dest').agg({'arr_delay':'sum'})
```

```python
# Find the worst days in terms of average arrival delay greater than 1 hour.
flights.groupby(['year','month','day']).agg({'arr_delay':'mean'}).query('arr_delay>60').sort_values('arr_delay').tail(10)
```

```python
# Find the number of carriers between each pair of origin and destination.
flights.groupby(['origin','dest']).agg({'carrier':'nunique'}).reset_index()

#alt way
flights.groupby(['origin','dest']).carrier.nunique()
```
