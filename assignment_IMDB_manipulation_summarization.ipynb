## Part 2: Data Summarization

```python
# Total "Gross" of all top 1000 movies
imdb['Gross'].sum()
```

```python
# Average "No_of_Votes" of all movies
imdb['No_of_Votes'].mean()
```

```python
# Count movies in each decade (e.g., ..., 1980, 1990, 2000, 2010, 2020)
# Sort decades by the number of movies in descending order
imdb['Decade'].value_counts().sort_values(ascending=False)
```

```python
# Count movies by different directors.
# Show the top 10 directors with the most movies in this list
imdb['Director'].value_counts().sort_values(ascending=False).head(10)
```

```python
# For each director, calculate the number of movies, average IMDB_Rating, and total gross.
# Ranked in descending order of total gross
# Show the top 10 directors with the most gross
(
    imdb.groupby('Director')  #groupby director name
    .agg({'Series_Title':'count',
          'IMDB_Rating':'mean',
          'Gross':'sum'})  #perform aggregation on each column seperately
    .sort_values(by= 'Gross', ascending=False).head(10)  #sort desc by gross
    )
```

```python
# Group movies by decade and director
# In each group (i.e., for each decade and each director),
# calculate the number of movies and average IMDB rating.
# Sort in descending order of movie count

imdb.groupby(['Decade', 'Director']).agg({'Series_Title':'count','IMDB_Rating':'mean'}).sort_values(by='Series_Title',ascending=False)
```

```python
# Find the top 3 highest rated movies of each year since 2010. - 'Ranking' column
# Show their released year, ranking, title, and IMDB_Rating.
# Sort them in descending order of year and ascending order of ranking

# must calculate a 'Ranking' column for each movie within its decade, based on IMDB_Rating -
                  #I asked Gemini how to get the top 3 by decade since I couldn't find in my notes how to return that and store it as a new column
imdb['Ranking'] = imdb.groupby('Released_Year')['IMDB_Rating'].rank(method='first', ascending=False) # it's grouping by Released_Year and then pulling IMDB_Ranking to do the .rank

# Now, filter for movies that are in the top 3 of their decade and are from decades since 2010
# Then select the desired columns and sort them as requested
( imdb[
    (imdb['Ranking'] <= 3) & (imdb['Released_Year'] >= 2010)
][
    ['Released_Year', 'Ranking', 'Series_Title', 'IMDB_Rating']
].sort_values(by=['Released_Year', 'Ranking'], ascending=[False, True])
)
```

```python
# Find all directors whose movies appeared in at least five different decades.
# Your result should include: director, decade, and the number of movies in the decade.

# Group by Director and Decade to get movie count for each combination
decade_movie_count = imdb.groupby(['Director', 'Decade']).size().reset_index(name='Movie_Count')

# Find directors with movies in at least 5 different decades
directors_with_five = decade_movie_count.groupby('Director')['Decade'].nunique()
directors_with_five = directors_with_five[directors_with_five >= 5].index

# Filter the decade_movie_counts to only include these directors
final_result = decade_movie_count[decade_movie_count['Director'].isin(directors_with_five)]

final_result.sort_values(by=['Director', 'Decade'])
```
